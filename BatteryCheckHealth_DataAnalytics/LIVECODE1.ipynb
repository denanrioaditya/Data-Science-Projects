Kamu adalah seorang data scientist yang bekerja di salah satu perusahaan manufaktur baterai Lithium ion. (Hint: Gunakan for loop dan if statement untuk mengerjakan perintah nomor b, kamu bisa menggunakan list untuk menyimpan perhitungan sementara, setelahnya dapat dimasukkan ke kolom R/P).

````python
#  Dua fungsi eksplisit masing-masing memuat perhitungan hambatan (R) dan daya (P) 
def hambatan(v,i):
    #Rumus menghitung hambatan
    return v/i

def daya(v,i):
   #Rumus menghitung daya dalam watt
    return v*i
    
````

````python
df.Cap.unique() #Melihat Value unique pada kolom Cap
````

````python
for cap in df.Cap.unique():
    if cap >= 2.5: #Validasi nilai cap diatas 2.5
         df['R/P'] = df.apply(lambda df: daya(df['V_m'], df['Current (A)']), axis=1) #Jika nilai cap >= 2.5
    else: 
        df['R/P'] = df.apply(lambda df: hambatan(df['V_m'], df['Current (A)']), axis=1) #Jika nilai cap < 2.5
````

````python
# rata-rata nilai kapasitas untuk arus = 1 Ampere dan cycle ganjil.
x=df.Cap[(df['Current (A)'] == 1)&(df['Cycle']%2 != 0)].mean()
print ('rata-rata nilai kapasitas untuk arus = 1 Ampere dan Cycle Ganjil adalah: ',x)
````
